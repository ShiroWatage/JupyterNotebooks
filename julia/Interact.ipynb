```julia
function logistic(a, num)
    u0 = 0.01
    u = u0
    u_sq = [ u0 ]
    
    for i in 1:num
        u = a*u*(1-u)
        push!(u_sq, u)
    end
    
    return u_sq
end
```

```julia
plot( logistic(1.2, 50), marker = :circle )
```

```julia
@manipulate for a in 0:0.01:2
    plot( logistic(a,100), marker = :circle)
end
```

```julia
import Colors
using Plots

function mycolorpicker()
    r = slider(0:255, label = "red")
    g = slider(0:255, label = "green")
    b = slider(0:255, label = "blue")
    output = Interact.@map Colors.RGB(&r/255, &g/255, &b/255)
    plt = Interact.@map plot(sin, marker=(3, :auto),color = &output)
    wdg = Widget(["r" => r, "g" => g, "b" => b], output = output)
    @layout! wdg hbox(plt, vbox(:r, :g, :b)) ## custom layout: by default things are stacked vertically
end
mycolorpicker()
```

```julia
function mycolorpicker2()
    r = slider(0:255, label = "red")
    g = slider(0:255, label = "green")
    b = slider(0:255, label = "blue")
    update = button("Update plot")
    output = Interact.@map (&update; Colors.RGB(r[]/255, g[]/255, b[]/255))
    plt = Interact.@map plot(sin, color = &output, marker=(3, :auto))
    wdg = Widget(["r" => r, "g" => g, "b" => b, "update" => update], output = output)
    @layout! wdg hbox(plt, vbox(:r, :g, :b, :update)) ## custom layout: by default things are stacked vertically
end
mycolorpicker2()
```

## Observables.@map マクロ
AbstractObservableが更新されるたびに計算がされなおされます

```julia
let 
    a = Observable(2)

    b = Observable(3)
    
    c = Observables.@map &a + &b
    println(c[])
    
    println("b becomes 100")
    b[] = 100
    println(c[]) # bが100になったことを受けて値が再計算されている
end
```

```julia
let 
    sld = slider(1:100)
    c = Interact.@map 1 + &sld |> println
    on(println, sld)
    sld
end
```
