```julia
function main()
    X = -2π:0.01:2π

    Y1 = sin.(X)
    Y2 = (X + sin.(X))/2
    Y3 = (X - sin.(X))
    
    plot(X, X, label="x")
    plot!(X, Y1, label="sin(x)")
    plot!(X, Y2, label="(x + sin(x))/2")
    plot!(X, Y3, label="x-sin(x)", legend=:bottomright)
end
```

```julia
main()
```

```julia
function func2()
    x = 0:0.01:π
    f(x) = min(x*π/2, 1 - cos(x*π/2))
    
    plot(x, f.(x))
    plot!(x, (x.-1)π/2 .+1)
end
func2()
```

```julia
function func3()
    x1 = -π:0.01:π
    x = -1:0.01:√2
    
#     plot(x, x  .- sin.(π/2 *x) )
    plot(x, √2 .- sqrt.(2 .- x.^2) )
    plot!(x1, 1 .- cos.(x1))
    plot!(x1, x1 )
end

func3()
```
